```python
len1 = 0
for cols in df['user']:
    len1+= len(cols)

avg_len1 = len1 / len(df['user'])
print(avg_len1)
```

```python
len2 = 0
for cols in df['assistant']:
    len2 += len(cols)

avg_len2 = len2 / len(df['assistant'])
print(avg_len2)
```

```python
labels = ["user query avg len", "assistant avg answer len"]
values = [avg_len1, avg_len2]
plt.bar(labels, values, color = ['skyblue', 'lightgreen'])
plt.ylabel("Average Length (characters)")
plt.title("Average Length of Queries vs Answers")
plt.show()
```

```python

all_topics = pd.concat([df["topic_1"], df["topic_2"], df["topic_3"]])
topic_counts = all_topics.value_counts().head(15)

topic_counts.plot(kind="barh", figsize=(10,7), color="orange")
plt.title("Top 15 Topics Across topic_1, topic_2, topic_3")
plt.xlabel("Count")
plt.ylabel("Topic Name")
plt.gca().invert_yaxis()
plt.show()
```

```python
topic_counts = df["topic_1"].value_counts().head(10)

topic_counts.plot(kind="barh", figsize=(10,6), color="skyblue")
plt.title("Top 10 Topics in Topic 1 Column")
plt.xlabel("Count")
plt.ylabel("Topic Name")
plt.gca().invert_yaxis()
plt.show()
```
